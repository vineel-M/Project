**CREATING A COUNT PLOT**:
 A count plot is a type of bar plot that displays the number of occurrences of each category in a categorical variable. In this case, the "PowerTrain" column is a categorical variable that contains different types of powertrains for vehicles

```python


def power_train(dataframe):
    plt.figure(figsize=(10,8))
    ax = sns.countplot(x= dataframe['PowerTrain'])
    plt.title('Count Plot of Powertrain', fontsize = 20)
    plt.xlabel('Power Train', fontsize = 15)
    plt.ylabel('Count', fontsize = 15)

    # Display count of each bar
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

    plt.show()
```

**Distribution of Body Styles by Powertrain Type**

```python

def bodystyle(dataframe):
    plt.figure(figsize=(12, 6))
    ax = sns.countplot(x='BodyStyle', data=dataframe, hue='PowerTrain')
    plt.title('Count plot of Body Style by Power Train', fontsize=20)
    plt.xlabel('Body Style', fontsize=15)
    plt.ylabel('Count', fontsize=15)

    # Set legend
    plt.legend(title='Power Train', loc='best', fontsize=12)

    # Display count of each bar
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points')

    # Add grid lines
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.show()
```

```python
bodystyle(df)
```

**INTERPRETATION:**
The business implication could be to gain insights into the popularity of different body styles for vehicles, as well as which powertrain types are preferred for each body style. This information could be used to inform marketing and product development decisions for car manufacturers, such as which body styles and powertrain types to focus on producing, or which marketing campaigns to target towards certain demographics that prefer specific body styles or powertrain types.

**CORRELATION MATRIX: HEATMAP**

```python
plt.figure(figsize=(15, 8))
sns.heatmap(df.corr(), linewidths=1, linecolor='white', annot=True)
plt.title('Correlation between features')
plt.show()
```

**INTERPRETATION:**
Heat map is a two dimensional plot of the amount of correlation (measure of dependence) between variables represented by colors. The varying intensity of color represents the measure of correlation. Correlation is a measure of linear relationship between two variables.
If the value is +1 or close to it then we say the variables are positively correlated.
 If the value is -1 or close to it then we say the variables are negatively correlated.
 If the value is 0 or close to it then we say the variables are not correlated.
